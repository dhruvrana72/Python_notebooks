```python
writer = pd.ExcelWriter( 'test.xlsx', engine='xlsxwriter')
for merchants in Merchant:
    df_merc = data_copy.loc[data_copy.TERMINALID == str(merchants)]
    df1 = pd.DataFrame()
    df1['totalamount'] = df_merc['totalamount']
    df1['DMRamount'] = df_merc['DMRamount']
    df1['NonDMRAmount'] = df1['totalamount'] - df1['DMRamount']
    df1['nonDMRcommission'] = df_merc['totalcommission']
    df1['DMRcommission_new'] = 0.01 * df_merc['DMRamount']
    df1['Totalcommission'] = df1['nonDMRcommission'] + df1['DMRcommission_new']
    df1['DMRcommission_old_given'] = df_merc['DMRcommission']
    df1.index = df_merc['date']
    df1 = df1[(df1.T != 0).any()]
    g = df1.groupby(pd.TimeGrouper("M"))
    DFList = [group[1] for group in g]
    test_df = pd.DataFrame()
    for dataframe in DFList:
        test_df = test_df.append(dataframe.groupby(pd.TimeGrouper(freq='M')).sum())
        #writer = pd.ExcelWriter( str(merchants) + '.xlsx', engine='xlsxwriter')
        test_df.to_excel(writer, sheet_name=str(merchants))
        #workbook  = writer.book
        #worksheet = writer.sheets[str(merchants)]
        #writer.save()
writer.save()
```

```python
from pandas import ExcelWriter
def dfs2xlsx(list_dfs,xls_path = None):
    #save_xls([df1,df2],'output1.xlsx')
    if xls_path == None :
        xls_path = 'rest_of_all.xlsx'
    writer = ExcelWriter(xls_path)
    i=2
    for n, df in enumerate(list_dfs):
        df.to_excel(writer,'rest_of_all',startcol=0,startrow =i)
        i+= len(df.index) + 2
    writer.save()
```
